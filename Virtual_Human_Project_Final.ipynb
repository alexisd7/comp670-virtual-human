```python
from llama_index.core import VectorStoreIndex, Settings
from llama_index.vector_stores.qdrant import QdrantVectorStore
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.llms.groq import Groq
from qdrant_client import QdrantClient

# Set up embedding model
embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/all-MiniLM-L6-v2")
Settings.embed_model = embed_model

# Set up Groq LLM
llm = Groq(
    model="llama3-8b-8192",
    api_key="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",
    temperature=0.7,
    api_url="https://api.groq.com/openai/v1"
)
Settings.llm = llm
Settings.streaming = False
Settings.num_output = 512
Settings.context_window = 3900

# Qdrant vector store
qdrant_client = QdrantClient(
    url="https://c4d519e8-841b-491e-8ffd-a70f8ab07f22.us-east-1-0.aws.cloud.qdrant.io",
    api_key="eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJhY2Nlc3MiOiJtIn0.sSiOCw3TUcPi_DmrmRPdruESddRctx1kYBVk8E3voGI"
)
vector_store = QdrantVectorStore(client=qdrant_client, collection_name="franklin_bot")
index = VectorStoreIndex.from_vector_store(vector_store)

# Create chat engine
chat_engine = index.as_chat_engine(chat_mode="context", llm=llm, verbose=True)

# Test queries
response = chat_engine.chat("What academic programs does Franklin University offer?")
print("Response 1:\n", response.response)

response = chat_engine.chat("What classes can be taken as part of the Computer Science program?")
print("\nResponse 2:\n", response.response)

response = chat_engine.chat("Can you tell me anything about the University of Central Florida?")
print("\nResponse 3:\n", response.response)
```
